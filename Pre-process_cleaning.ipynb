### **Terminology** - Term, Document, Corpus

**1.Term/Token-term** - a single/group of words together is a term.



**2.Document** - Collection of terms



- If 1 pdf, 1 paragraph can be considered as a document.
- If multiple pdf's/text files then each pdf/text file can be considered as a document.

One frequently used lemmatizer is Wordnet. It is a database of dictionary/words. It is a database of most commonly used words will have its corresponding root form i.e the lemma of that word.



**Difference b/w Stemming & Lematization**

1. Stemming is rule based while Lemmatization is dictionary/lookup based.

Stemming is a process that stems or removes last few characters from a word, often leading to incorrect meanings and spelling. It is rule based, like always remove 'ing' or 'ly' from end of word. For instance, stemming the word 'Caring' would return 'Car', organisation would return 'organ'. Lemmatization considers the context and converts the word to its meaningful base form, which is called Lemma. 

2. Stemming can be preferred if speed is required for large number of documents, lemmatization if human readable output required after identifying root form of word.

3. For Stem u do not need parts of speech of word while for lemmatization you need parts of speech of each word.

## 3. Stop Words removal

Commonly occuring words might not be useful for most text mining analysis. Can be removed from dataset.



- NLTK library has around 170 stop words identified for the English language.

In Sentiment analysis we might not remove stop words or be careful while removing stop words. e.g. if we remove stop word 'not' then -ve sentiment is interpreted as +ve sentiment instead.
